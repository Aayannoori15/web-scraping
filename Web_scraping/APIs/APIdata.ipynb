```python
import requests 
```

```python
headers={
    'User-Agent':'Mozilla/5.0'
}
url='https://www.scrapethissite.com/pages/simple/'
res=requests.get(url,headers=headers)
```

```python
with open('Scrapped_data/data1.html','w') as f:
    f.write(res.text)
```

```python
#beautifulsoup
```

```python
from bs4 import BeautifulSoup
```

```python
with open('Scrapped_data/data1.html','r') as f:
    html_content=f.read()
```

```python
soup=BeautifulSoup(html_content,'lxml')
```

```python
#getting data 
```

```python
country_list=soup.find_all('h3')
df_country=[]
df_capital=[]
df_population=[]
for countries in country_list:
    df_country.append(countries.get_text().strip())
    df_capital.append(countries.find_next('div').select('span.country-capital')[0].get_text())
    df_population.append(countries.find_next('div').select('span.country-population')[0].get_text())
```

```python
#importing pandas to make dataframe of the information fetched 
```

```python
import pandas as pd 
```

```python
country_data={'country':df_country,'capital':df_capital,'popultion':df_population}
df=pd.DataFrame(country_data)
```

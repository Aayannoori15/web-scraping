```python
import requests 
from bs4 import BeautifulSoup
```

```python
page_count=1
while True:
    url=f'https://quotes.toscrape.com/page/{page_count}/'
    res=requests.get(url)
    soup=BeautifulSoup(res.text,'lxml')
    quotes=soup.select('div.quote')
    if not quotes:
        print('no more pages')
        break
    else:
        with open(f'Scrapped_data/activity1_data{page_count}.html','w') as f:
            f.write(res.text)
        print(f'page{page_count} downloaded')
        page_count+=1
```

extracting useful information

```python
life_quotes=[]
for page_count in range(1,11):
    with open (f'Scrapped_data/activity1_data{page_count}.html','r') as f:
        html_content=f.read()
    soup=BeautifulSoup(html_content,'lxml')
    quotes=soup.select('div.quote')
    for q in quotes:
        all_tags=q.select('.tags .tag')
        for t in all_tags:
            if 'life' == t.get_text():
                life_quotes.append(q.select_one('span.text').get_text())
                life_quotes.append(q.select_one('small.author').get_text())
                print(life_quotes)
                life_quotes=[]
    print(f'********Page{page_count}over*********')
```
